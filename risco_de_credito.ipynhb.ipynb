```python
import pandas as pd
import numpy as np

df = pd.read_csv("UCI_Credit_Card.csv")
```

```python
df = df.rename(columns={'default.payment.next.month': 'default'})
```

```python
df['EDUCATION'] = df['EDUCATION'].replace({0: 4, 5: 4, 6: 4})
```

```python
df['MARRIAGE'].unique()
```

```python
df['MARRIAGE'] = df['MARRIAGE'].replace({0: 3})
```

```python
bill_columns = ['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']
df['AVG_BILL'] = df[bill_columns].mean(axis=1)
```

```python
df['LIMIT_UTILIZATION'] = df['AVG_BILL'] / (df['LIMIT_BAL'] + 1)
```

```python
pay_columns = ['PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5']
df['AVG_PAYMENT'] = df[pay_columns].mean(axis=1)
df['PAYMENT_TO_BILL_RATIO'] = np.where(df['AVG_BILL'] > 0, df['AVG_PAYMENT'] / df['AVG_BILL'], 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
```

```python
X = df.drop(columns=['ID', 'default'])
y = df['default']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
numeric_features = [
    'LIMIT_BAL', 'AGE', 'AVG_BILL', 'LIMIT_UTILIZATION', 'PAYMENT_TO_BILL_RATIO',
    'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
    'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'
]

categorical_features = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(handle_unknown='ignore', drop='first'), categorical_features)
])
```

```python
scale_pos_weight = (len(y_train) - sum(y_train)) / sum(y_train)
```

```python
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(
        scale_pos_weight=scale_pos_weight,
        random_state=42,
        eval_metrics='logloss',
        n_estimators=100,
        max_depth=5,
        learning_rate=0.1
    ))
])
```

```python
print("Training the model...")
model_pipeline.fit(X_train, y_train)
print("Model trained successfully!")
```

```python
y_pred = model_pipeline.predict(X_test)
y_pred_proba = model_pipeline.predict_proba(X_test)[:,1]

print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred_proba):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_encoder = model_pipeline.named_steps['preprocessor'].named_transformers_['cat']
encoded_cat_features = list(cat_encoder.get_feature_names_out(categorical_features))

all_features = numeric_features + encoded_cat_features

importances = model_pipeline.named_steps['classifier'].feature_importances_

feature_importance_df = pd.DataFrame({
    'Feature': all_features, 
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(
    x='Importance',
    y='Feature',
    data=feature_importance_df.head(15),
    palette='viridis'
)
plt.title('Top 15 Most Important Features - XGBoost Credit Model', fontsize=14, fontweight='bold')
plt.xlabel('Feature Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
cm = confusion_matrix(y_test, y_pred)

print("--- TEST RESULTS ---")
print(f"Total number of customers tested: {len(X_test)}")
print(f"Correctly predicted non-defaults (True Negatives): {cm[0, 0]}")
print(f"Correctly predicted defaults (True Positives): {cm[1, 1]}")
print(f"Predicted default, but the customer was not in default (False Positives): {cm[0, 1]}")
print(f"Failed to detect a default (False Negatives): {cm[1, 0]}")
```
